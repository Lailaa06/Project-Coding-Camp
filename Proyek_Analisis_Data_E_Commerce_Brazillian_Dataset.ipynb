## Analisis Lanjutan (Opsional)

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
orders = pd.read_csv("/content/drive/My Drive/orders_dataset.csv")
order_items = pd.read_csv("/content/drive/My Drive/order_items_dataset.csv")
payments = pd.read_csv("/content/drive/My Drive/order_payments_dataset.csv")
customers = pd.read_csv("/content/drive/My Drive/customers_dataset.csv")

# Menghubungkan payments dengan orders untuk mendapatkan customer_id
payments = payments.merge(orders[['order_id', 'customer_id']], on='order_id', how='left')

# Mengonversi tanggal pembelian menjadi format datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
rfm_data = orders[['customer_id', 'order_id', 'order_purchase_timestamp']]

# Recency: Menghitung jumlah hari sejak pembelian terakhir
max_date = rfm_data['order_purchase_timestamp'].max()
recency = rfm_data.groupby('customer_id').order_purchase_timestamp.max().reset_index()
recency['Recency'] = (max_date - recency['order_purchase_timestamp']).dt.days

# Frequency: Menghitung jumlah transaksi per pelanggan
frequency = rfm_data.groupby('customer_id').order_id.nunique().reset_index()
frequency.columns = ['customer_id', 'Frequency']

# Monetary: Menghitung total pengeluaran pelanggan
monetary = payments.groupby('customer_id').payment_value.sum().reset_index()
monetary.columns = ['customer_id', 'Monetary']

# Menggabungkan semua metrik RFM
df_rfm = recency.merge(frequency, on='customer_id').merge(monetary, on='customer_id')

# Membuat skor RFM berdasarkan kuartil
df_rfm['R_Score'] = pd.qcut(df_rfm['Recency'], q=4, labels=[4, 3, 2, 1])
df_rfm['F_Score'] = pd.qcut(df_rfm['Frequency'].rank(method='first'), q=4, labels=[1, 2, 3, 4])
df_rfm['M_Score'] = pd.qcut(df_rfm['Monetary'], q=4, labels=[1, 2, 3, 4])

df_rfm['RFM_Score'] = df_rfm[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)

# Klasifikasi pelanggan berdasarkan skor RFM
def rfm_segment(score):
    if score >= 10:
        return 'Best Customers'
    elif score >= 8:
        return 'Loyal Customers'
    elif score >= 6:
        return 'Potential Loyalist'
    elif score >= 4:
        return 'At Risk'
    else:
        return 'Lost Customers'

df_rfm['Segment'] = df_rfm['RFM_Score'].astype(int).apply(rfm_segment)

# Visualisasi distribusi RFM
plt.figure(figsize=(12,5))
sns.histplot(df_rfm['Recency'], bins=30, kde=True)
plt.title('Distribusi Recency')
plt.show()

plt.figure(figsize=(12,5))
sns.histplot(df_rfm['Frequency'], bins=30, kde=True)
plt.title('Distribusi Frequency')
plt.show()

plt.figure(figsize=(12,5))
sns.histplot(df_rfm['Monetary'], bins=30, kde=True)
plt.title('Distribusi Monetary')
plt.show()

# Menampilkan hasil akhir
print(df_rfm.head())
```

### Insight

*   Diagram Distribusi Recency
1.  Banyak pelanggan yang baru belanja (<100 hari), ini pelanggan aktif.
2. Banyak juga yang lama tidak belanja (>500 hari), ini pelanggan yang mungkin sudah "hilang."
3. Bisa targetkan ulang pelanggan lama dan pertahankan yang baru dengan strategi loyalitas.

* Diagram Distribusi Frequency
1. Banyak pelanggan yang hanya belanja sekali dan tidak kembali lagi.
2. Strategi retensi pelanggan perlu diperbaiki untuk meningkatkan repeat purchases.
3. Bisa jadi ada faktor seperti harga, pengalaman belanja, atau produk yang membuat pelanggan tidak melakukan transaksi lagi.

* Diagram Distribusi Monetary
1. Mayoritas pelanggan bertransaksi kecil, sedikit yang bertransaksi besar (outliers).  
2. Banyak pelanggan masuk kategori *Potential Loyalist*, tapi ada yang *At Risk*.  
3. Fokus pada retensi pelanggan bernilai tinggi & reaktivasi pelanggan berisiko.
